```python
df = df.drop_duplicates()
df.shape
```

```python
#Average final grade (G3).
avg_g3 = df['G3'].mean()
avg_g3
```

```python

# How many students scored above 15?
above_15 = df[df['G3'] > 15].shape[0]
above_15
```

```python
#Is study time correlated with performance?
df[['studytime', 'G3']].corr()
```

```python
#Which gender performs better on average?
df.groupby('sex')['G3'].mean()
```

```python
#Histogram of grades.
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
#Scatterplot: study time vs grades.
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
#Bar chart: male vs female average score.

gender_avg = df.groupby('sex')['G3'].mean()
gender_avg

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
gender_avg.plot(kind='bar')

plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.title("Average Final Grade by Gender")

plt.show()
```
